```python
import pandas as pd
import numpy as np

# Assuming 'df' is the DataFrame you created from the Excel file

# Calculate the distance matrix
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Selecting the relevant columns
coordinates = df[['X - coordinate', 'Y - coordinate']].values

num_nodes = len(coordinates)
distance_matrix = np.zeros((num_nodes, num_nodes))

for i in range(num_nodes):
    for j in range(num_nodes):
        x1, y1 = coordinates[i]
        x2, y2 = coordinates[j]
        distance_matrix[i, j] = euclidean_distance(x1, y1, x2, y2)

# Convert distance matrix to DataFrame for better visualization (optional)
distance_df = pd.DataFrame(distance_matrix, columns=df['Node'], index=df['Node'])

# Print the distance matrix
(distance_df)
```

```python
#!pip install pulp

import pulp

# Define problem
problem = pulp.LpProblem("Logistics_Problem", pulp.LpMinimize)

# Define decision variables
n = len(df)
Q = 90

x = {(i, j): pulp.LpVariable(f"x_{i}_{j}", lowBound=0, cat=pulp.LpInteger) 
     for i in range(1, n+1) for j in range(1, n+1)}

# Construct the distance matrix
c = {(i, j): ((df.loc[i-1, 'X - coordinate'] - df.loc[j-1, 'X - coordinate'])**2 + 
              (df.loc[i-1, 'Y - coordinate'] - df.loc[j-1, 'Y - coordinate'])**2)**0.5
     for i in range(1, n+1) for j in range(1, n+1)}

# Define objective function
problem += pulp.lpSum(c[i, j] * x[i, j] for i in range(1, n+1) for j in range(1, n+1))

# Demand Constraints
for i in range(2, n+1):
    problem += pulp.lpSum(x[i, j] for j in range(1, n+1)) == df.loc[i-1, 'Demand, MT']

# Supply Constraint
problem += pulp.lpSum(x[i, 1] for i in range(2, n+1)) == Q

# Capacity Constraints
for i in range(1, n+1):
    problem += pulp.lpSum(x[i, j] for j in range(1, n+1)) <= Q

# Solve the problem
problem.solve()

# Print the results
for i in range(1, n+1):
    for j in range(1, n+1):
        if pulp.value(x[i, j]) != 0:
            print(f"Transport {pulp.value(x[i, j])} MT from Node {i} to Node {j}")

print(f"Total Cost: {pulp.value(problem.objective)}")
```

```python
import pandas as pd
import pulp
import matplotlib.pyplot as plt


edges = []

for i in range(1, n+1):
    for j in range(1, n+1):
        if pulp.value(x[i, j]) != 0:
            edges.append((i, j))

# Print the results
for i, j in edges:
    print(f"Transport {pulp.value(x[i, j])} MT from Node {i} to Node {j}")

print(f"Total Cost: {pulp.value(problem.objective)}")

# Visualize the graph
plt.figure(figsize=(8, 8))
plt.scatter(df['X - coordinate'], df['Y - coordinate'], color='blue', s=100)
for i, row in df.iterrows():
    plt.annotate(f'Node {int(row["Node"])}', (row['X - coordinate']+1, row['Y - coordinate']))

for i, j in edges:
    plt.plot([df.loc[i-1, 'X - coordinate'], df.loc[j-1, 'X - coordinate']], 
             [df.loc[i-1, 'Y - coordinate'], df.loc[j-1, 'Y - coordinate']], 'k--', lw=1)

plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Logistics Network')
plt.grid(True)

plt.show()
```
